Formulation of headings and normal text

```python
document = Document(todocx)

headings = []
texts = []
para = []
for paragraph in document.paragraphs:
    if paragraph.style.name == "Heading 1":
        if headings:
            texts.append(para)
        headings.append(paragraph.text)
        para = []
    elif paragraph.style.name == "Normal":
        para.append(paragraph.text)
if para or len(headings)>len(texts):
    texts.append(texts.append(para))

zipped = list(zip(headings,texts))
zipped
```

Converting from list to dataframe object 

```python
df = DataFrame (zipped,columns=['Bold_text', 'Normal_text']) #Converting text to Dataframe

df['Normal_text'] = [','.join(map(str, l)) for l in df['Normal_text']]

df = pd.concat([Series(row['Bold_text'], row['Normal_text'].split('.,'))  # Split the normal text            
                    for _, row in df.iterrows()]).reset_index() 
df.columns = ['Normal_text', 'Bold_text']

df['Country_Bold'] = None
for item in ['EU', 'Germany','German', 'Hungary', 'Spanish', 'Italy', 'Swiss', 'European','Nordic','Belgium','Polish','Spain'
            ,'Denmark','UK','Greek','Japan','Portugal','Italian','Europe','Czech','French','Netherlands','Dutch','Dubai',
            'Finland','Austrian','UAE','INDIA','Greek','CEZ','Poland','France','Turkey','Portuguese','Scotland','Netherlands'
            ,'Great Britain','Norway','Sweden','Russia','Lithuania','Switzerland']: # Add new country _ Bold text
    df['Country_Bold'][df['Bold_text'].str.contains(item)] = item
    
repl_dict = {"European": "European Union","EU": "European Union","German": "Germany","Swiss": "Switzerland","Spanish": "Spain",
            "UK": "United Kingdom","UK's": "United Kingdom","Japan": "Japan","UK": "United Kingdom","Italian": "Italy",
            "Europe": "European Union","Poland": "Poland","Portuguese": "Portugal","French": "France","Polish": "Poland",
             "Russian": "Russia","Dutch": "Netherland"} # Add Country mapping _ Bold text
df['Country_Bold'] = df['Country_Bold'].replace(repl_dict)

df['Country_Normal'] = None
for item in ['EU', 'Germany','German', 'Hungary', 'Spanish', 'Italy', 'Swiss', 'European','Nordic','Belgium','Polish','Spain'
            ,'Denmark','UK','Greek','Japan','Portugal','Italian','Europe','Czech','French','Netherlands','Dutch','Dubai',
            'Finland','Austrian','UAE','INDIA','Greek','CEZ','Poland','France','Turkey','Portuguese','Scotland','Netherlands'
            ,'Great Britain','Norway','Sweden','Russia','Lithuania','India']: # Add new country _ Normal text
    df['Country_Normal'][df['Normal_text'].str.contains(item)] = item
    
repl_dict = {"European": "European Union","EU": "European Union","German": "Germany","Swiss": "Switzerland","Spanish": "Spain",
            "UK": "United Kingdom","UK's": "United Kingdom","Japan": "Japan","UK": "United Kingdom","Italian": "Italy",
            "Europe": "European Union","Poland": "Poland","Portuguese": "Portugal","Polish": "Poland","French": "France",
            "Russian": "Russia","Dutch": "Netherland"} # Add Country mapping _ Normal text
df['Country_Normal'] = df['Country_Normal'].replace(repl_dict)

item = ['Nord Pool','NP', 'Epex Spot','EPEX SPOT', 'Epex', 'EXAA', 'NASDAQ','N2EX', 'OMIE', 'SEMO', 'SEMOpx', 
        'GME','EEX','BSP','CROPEX','HEnEx','HUPX','eex spot','IBEX', 'OKTE', 'OPCOM', 'OTE','TGE','EPEX']
df['competitor'] = df['Normal_text'].str.findall('|'.join(item)).apply(set).str.join(', ') # Competitor extraction 

item = ['Long-term', 'year ahead', 'front year', 'month ahead', 'CAL', 'year', 'quarter', 'front quarter','Q1','Q2','Q3','Q4', 
        'front month', 'week ahead', 'front week', 'CW', 'day-ahead', 'spot', 'short term', 'intraday','Intraday','intra-day']
df['Market'] = df['Normal_text'].str.findall('|'.join(item)).apply(set).str.join(', ') # Market extraction

date_reg = r'.*([1-3][0-9]{3})'
df['Date_Filter'] = df['Bold_text'].str.extract(date_reg, expand=False) # Year extraction

item = ['lift','lowest','lower','fall','price','higher','cut','climb','increase','High','decline','drop','rise',
        'decrease','dip','down','low','rebound','loss','gain','negative','positive','returns']
df['Highlight'] = df['Bold_text'].str.findall('|'.join(item)).apply(set).str.join(', ') # Highlights extraction

item = ['Coal', 'Natural gas', 'oil', 'Nuclear', 'Hydro', 'Solar','solar', 'Wind', 'Biomass', 'power', 'CO2'
        , 'carbon', 'hydrogen','wind','temperature','temp']
df['Commodity'] = df['Normal_text'].str.findall('|'.join(item)).apply(set).str.join(', ') # Commodity extraction

file_name = Doc_file
file_name_ending = file_name.rsplit('_',1)[-1]
df['Date'] = file_name_ending.rsplit('.',1)[0] #Date of publication

stop = stopwords.words('english') #Stopwords 
df['tweet_without_stopwords'] = df['Bold_text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))

df['Highlight'].replace('', np.nan, inplace=True)
df['Commodity'].replace('', np.nan, inplace=True)
df['competitor'].replace('', np.nan, inplace=True)
df['Market'].replace('', np.nan, inplace=True)
df['concat_series'] = df[['Highlight', 'Commodity', 'competitor', 'Market']].apply(lambda x: ', '.join(x[x.notnull()]), axis=1) #concatenate the series
df['comma_count'] = df['concat_series'].str.count('\w+')

df['ranked']=df['comma_count'].rank(ascending=0,method='dense') #Ranking
now = datetime.datetime.now().year
df['Date_Filter'] = pd.to_numeric(df['Date_Filter'], errors='coerce')

#Country Ranking
df['country_mask'] = df['Country_Bold'].isin(('Germany', 'United Kingdom','Poland','France','European Union'))
df['country_mask'] = np.where(df['country_mask'] == False, 9, df['country_mask'])
df['country_mask'] = np.where((df['Country_Bold'].isin(('Germany', 'United Kingdom','Poland','France','European Union'))) & (~df['Country_Normal'].isin(('Germany', 'United Kingdom','Poland','France','European Union'))),7, df['country_mask'])
df['country_mask'] = np.where(df['Date_Filter'] > now, 8, df['country_mask'])

df.to_excel('J:/13-data_management/05-Data Sets/28. Relevant Press articles/excel/' + attachment_name[:13] + '.xlsx') #Export Raw file in excel 
df.head()
```
